```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = [50, 60, 70, 80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190]  #area de casas
Y = [85000, 99000, 112000, 124000, 136000, 148000, 160000, 172000, 185000, 198000, 210000, 222000, 235000, 247000, 260000] #costos de casa
X_ones=np.c_[np.ones(15),X]
X_ones
```

```python
tita = np.linalg.inv(X_ones.T.dot(X_ones)).dot(X_ones.T).dot(Y)
tita
```

```python
plt.scatter(X,Y, s=40, c='#06d6a0')

X_extremo = [40,200]
X_extremos_ones = np.c_[np.ones(2),X_extremo]
Y_extremo = X_extremos_ones.dot(tita)
plt.plot(X_extremo,Y_extremo,'r-')

plt.axis([40,200,82000,300000])
plt.xlabel("Areas")
plt.ylabel("Costo")
plt.title("AreaXCosto")
plt.grid()
```

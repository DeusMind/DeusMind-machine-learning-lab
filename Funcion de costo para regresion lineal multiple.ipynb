```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion_cost(X, y, w, b):
    m = X.shape[0]
    cost = 0
    for i in range(0,m):
        f_wb_i = np.dot(w,X[i]) + b
        cost = cost + (f_wb_i - y[i])**2
    cost = cost/(2*m)
    return(np.squeeze(cost))

#Ejemplo 
# X: Matriz de características (m=2, n=3)
X = np.array([
    [2104, 5, 45],  # Casa 1
    [1416, 3, 40]   # Casa 2
])

# y: Vector de precios reales (lo que queremos predecir)
y = np.array([460, 232])

# w: Pesos iniciales (uno por cada característica)
w = np.array([0.1, 4.0, 1.0])

# b: Sesgo inicial
b = 10.5

costo = funcion_cost(X, y, w, b)
print(costo)
```

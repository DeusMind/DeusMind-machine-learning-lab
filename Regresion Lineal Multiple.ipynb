```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Vectorizado
def regresion_lineal_multiple_V(X, w, b):
    f_wb = np.dot(w,X) + b
    return(f_wb)

#No Vectorizado
def regresion_lineal_multiple(X,w,b):
    n = len(X)
    p = 0
    for i in range(0,n):
        p_i = w[i] * X[i]
        p = p + p_i
    p = p + b
    return(p)

X = np.array([1.2, 3, 0.5])
w = np.array([0.5, 0.2, 0.1])
b = 0.4

Y_predictora1 = regresion_lineal_multiple_V(X, w, b)
Y_predictora2 = regresion_lineal_multiple(X, w, b)
print(Y_predictora1)
print(Y_predictora2)
```

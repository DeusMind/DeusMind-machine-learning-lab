```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Vectorizado
def regresion_lineal_multiple_V(X, w, b):
    f_wb = np.dot(X,w) + b
    return(f_wb)

#No Vectorizado
def regresion_lineal_multiple(X,w,b):
    m = X.shape[0]  # Número de casas (filas)
    n = X.shape[1]  # Número de características (columnas)
    
    # Creamos una lista o array vacío para guardar el precio de cada casa
    predicciones = np.zeros(m)
    
    for i in range(m):          # Ciclo para cada CASA
        suma_caracteristicas = 0
        for j in range(n):      # Ciclo para cada CARACTERÍSTICA (peso * x)
            suma_caracteristicas += X[i, j] * w[j]
        
        predicciones[i] = suma_caracteristicas + b
        
    return predicciones

X = np.array([[1.2, 3, 0.5], [2, 4.5, 1]])
w = np.array([0.5, 0.2, 0.1])
b = 0.4

Y_predictora1 = regresion_lineal_multiple_V(X, w, b)
Y_predictora2 = regresion_lineal_multiple(X, w, b)
print(Y_predictora1)
print(Y_predictora2)
```
